```python
angles = np.arange(0,800,5)
angle_in_rand = np.radians(angles)
sin_values=np.sin(angle_in_rand)
cos_values=np.cos(angle_in_rand)
```

```python
plt.title("Sine Graph")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.xticks(np.arange(0,811,90))
plt.grid(True)
plt.plot(angles, sin_values, color='b',label='sin' )
plt.legend()
plt.show()
```

```python
plt.title("Sine Graph")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.xticks(np.arange(0,811,90))
plt.grid(True)
plt.plot(angles, cos_values, color='r', label='cos')
plt.legend()
plt.show()
```

```python
plt.title("Sine Graph")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.xticks(np.arange(0,811,90))
plt.grid(True)
plt.plot(angles, -sin_values, color='g', label='-sin')
plt.legend()
plt.show()
```

```python
plt.title("Sine Graph")
plt.xlabel("Angles")
plt.ylabel("Values")
plt.xticks(np.arange(0,811,90))
plt.grid(True)
plt.plot(angles, sin_values, color='b',label='sin' )
plt.plot(angles, cos_values, color='r', label='cos')
plt.plot(angles, -sin_values, color='g', label='-sin')
plt.legend()
plt.show()
```

```python
activity=['sleep','study','social Network', 'Play','Other Work']
hrs=[8,3,11,0.5, 1.5]
colour = ['b','r','g','m','y']
slice = (0.1, 0.1, 0.1, 0.5,0.13)  # only "explode" the slice
plt.pie(hrs, explode = slice, labels = activity, autopct='%.2f%%', colors=colour, shadow=True, startangle=45)
plt.show()
```

```python
days = [1,2,3,4,5]
Sleeping = [7,8,6,11,7]
Reading = [2,3,4,3,2]
Eating = [7,8,7,2,2]
Playing = [8,5,7,8,13]
plt.figure(figsize=(10,8))
# Plot empty lines for labels
plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Reading',linewidth=5)
plt.plot([],[],color='r',label='Eating',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)

plt.stackplot(days, Sleeping, Reading, Eating, Playing, colors = ['m','c','r','k'])
plt.xlabel('X axis ----->')
plt.ylabel('Y axis ----->')
plt.title('Stack Plot')
plt.legend()
plt.savefig('stack plot.png',dpi = 300, facecolor = 'w', box_inches = 'tight', transparent = False)
plt.show()
```
